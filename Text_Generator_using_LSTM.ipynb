```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
Tokenizer=Tokenizer()
```

```python
Tokenizer.fit_on_texts([faqs])
```

```python
sequences=[]
for sentence in faqs.split('\n'):
    tokenized_sent=Tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1, len(tokenized_sent)):
        n_gram=tokenized_sent[:i+1]
        sequences.append(n_gram)
```

```python
print(sequences)
```

```python
max_length=max([len(x) for x in sequences])
```

```python
print(max_length)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input=pad_sequences(sequences, maxlen=max_length, padding='pre')
```

```python
len(Tokenizer.word_index)
```

```python
X= padded_input[:,:-1] # We are picking every sequences except the last one
y= padded_input[:,-1]
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical

y=to_categorical(y, num_classes=1011)
```

Up till now we have made our training data

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model=Sequential()
# Input length 56 because we have 56 rows
model.add(Embedding(1011, 100, input_shape=(159,)))
model.add(LSTM(200))
model.add(Dense(1011, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```
